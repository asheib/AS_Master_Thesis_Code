```python
import pandas as pd
import matplotlib.pyplot as plt

# Define the path to the CSV file
file_path = r"E:\all folders\STUDIES\UZH\master thesis\IMIS data\new data\new processed data\TSS_30MIN_MEAN_files\Stations covering 2000till 2023\analysis\Filtered_Altitude_data.csv"

# Load the CSV file
df = pd.read_csv(file_path)

# Check the first few rows to understand the structure
print(df.head())

# Plot the altitude data
plt.figure(figsize=(15, 6))
plt.plot(df['Station'], df['Altitude'], marker='o', linestyle='-', color='b', markersize=5)
plt.xticks(rotation=90, fontsize= 10)
plt.yticks(fontsize= 15)
plt.xlabel('Station', fontsize= 15)
plt.ylabel('Altitude (m)', fontsize= 15)
plt.title('Altitude of 80 Ground Stations', fontsize=20)
plt.tight_layout()  # Adjust layout for readability
plt.show()
```
