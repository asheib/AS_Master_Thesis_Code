```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# Construct file path
base_path = r"E:\all folders\STUDIES\UZH\master thesis\IMIS data\new data\new processed data\TSS_30MIN_MEAN_files\elevation_aspect_slope"
file_name = "elevation-aspect-slope-80 stations.csv"
file_path = os.path.join(base_path, file_name)

# Verify file existence
if not os.path.exists(file_path):
    print(f"File not found: {file_path}")
else:
    # Load data
    data = pd.read_csv(file_path)

    # Ensure required columns
    if 'Station' in data.columns and 'aspect' in data.columns:
        # Plot
        data = data.sort_values(by='aspect', ascending=True)
        plt.figure(figsize=(16, 8))  # Increase figure width
        plt.bar(data['Station'], data['aspect'], color='skyblue', edgecolor='black', linewidth=1, width=0.8)
        plt.xlabel('Station', fontsize=20)
        plt.ylabel('Aspect', fontsize=20)
        plt.title('Aspect of Each Station', fontsize=30)
        plt.xticks(rotation=90, ha='right', fontsize=12)  # Rotate and align labels
        plt.yticks(fontsize=20)
        plt.tight_layout(pad=2)  # Add space around the plot
        plt.show()
    else:
        print("The required columns 'Station' and 'aspect' are not found in the file.")

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = r'E:\all folders\STUDIES\UZH\master thesis\MODIS DATA\Actual Data\MOD21A1N\AMD2.csv'
data = pd.read_csv(file_path)

# Ensure 'datetime' is in datetime format
data['datetime'] = pd.to_datetime(data['datetime'], errors='coerce')

# Filter data for the year 2000
data_2000 = data[data['datetime'].dt.year == 2000]

# Plot a histogram of the 'View_Angle_scaled' for the year 2000
plt.figure(figsize=(10, 6))
plt.hist(data_2000['View_Angle_scaled'].dropna(), bins=30, color='skyblue', edgecolor='black')

# Customize the plot
plt.title('Histogram of Frequencies of View Angle of Station AMD for the Year 2000', fontsize=18)
plt.xlabel('View Angle (Scaled)', fontsize=17)
plt.ylabel('Frequency', fontsize=17)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.tight_layout()

# Show the plot
plt.show()

```
