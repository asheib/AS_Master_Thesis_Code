```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the results from the saved CSV file
results_path = "E:\\all folders\\STUDIES\\UZH\\master thesis\\MODIS DATA\\Actual Data\\MOD21A1N\\Filtered_Data\\mean_temperature_trends_by_hour.csv"
results = pd.read_csv(results_path)

# Create a pivot table for the heatmap using the "Trend" column
heatmap_data = results.pivot(index="Hour", columns="Station", values="SlopePerYear")  # Switch index and columns

# Plotting the heatmap without annotations and with a green-white color palette
plt.figure(figsize=(25, 4))
heatmap = sns.heatmap(heatmap_data, cmap='Greens', annot=False, cbar_kws={"label": "Temperature Trend"},linewidths=0.5)

# Increase font sizes for title, labels, and ticks
plt.title("Mean MOD21A1N LSTemperature Trends by Hour for Each Station", fontsize=20)
plt.xlabel("Station", fontsize=18)
plt.ylabel("Hour of Day", fontsize=18)
plt.xticks(rotation=90, fontsize=14)  # Rotate x-axis labels for better visibility and increase size
plt.yticks(fontsize=14)  # Increase size for y-axis ticks

# Increase size for color bar ticks and label
colorbar = heatmap.collections[0].colorbar  # Get the color bar from the heatmap object
colorbar.ax.tick_params(labelsize=16)  # Increase size for color bar ticks
colorbar.set_label("Temperature Trend", fontsize=16)  # Set color bar label size

plt.tight_layout()
plt.show()


```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the results from the saved CSV file
results_path = "E:\\all folders\\STUDIES\\UZH\\master thesis\\MODIS DATA\\Actual Data\\MOD21A1N\\Filtered_Data\\mean_temperature_trends_by_hour.csv"
results = pd.read_csv(results_path)

# Create a pivot table for the table
table_data = results.pivot(index="Hour", columns="Station", values="SlopePerYear")  # Switch index and columns
table_data = table_data.round(2)  # Round to two decimal places

# Create a figure and axis with an even larger size for better visibility
fig, ax = plt.subplots(figsize=(50, 20))  # Large figure size

# Remove axis
ax.axis('tight')
ax.axis('off')

# Set a color palette
sns.set(style="whitegrid")
col_colors = ["#f7f7f7"] * len(table_data.columns)  # Light gray for column headers
row_colors = ['#f1f1f1' if i % 2 == 0 else 'white' for i in range(len(table_data))]  # Alternating row colors

# Plot the table with a designed look
table = ax.table(cellText=table_data.values,
                colLabels=table_data.columns,
                rowLabels=table_data.index,
                cellLoc='center',
                loc='center',
                colColours=col_colors,
                rowColours=row_colors,
                colWidths=[0.1] * table_data.shape[1])

# Customize font properties for readability
table.auto_set_font_size(False)
table.set_fontsize(20)  # Increase font size for better visibility
table.scale(0.2, 5)  # Increase table scaling for better spacing

# Bold the column and row labels
for (i, j), cell in table.get_celld().items():
    if i == 0:
        cell.set_fontsize(22)  # Larger font for header
        cell.set_text_props(weight='bold')
        cell.set_facecolor('#d3d3d3')  # Light gray background for headers
    elif j == 0:
        cell.set_fontsize(22)  # Larger font for row labels
        cell.set_text_props(weight='bold')
        cell.set_facecolor('#d3d3d3')  # Light gray for row labels

# Save the table to a file (e.g., PNG format)
table_data.to_csv(r'E:\all folders\STUDIES\UZH\master thesis\MODIS DATA\Actual Data\MOD21A1N\\Filtered_Data\Stations trends per hour.csv')

# Show the table
plt.show()
```
